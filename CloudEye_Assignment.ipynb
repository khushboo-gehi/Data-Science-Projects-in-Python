```python
#Feature correlation
from yellowbrick.target import FeatureCorrelation
visualizer = FeatureCorrelation(labels=cl)

visualizer.fit(x,y)        # Fit the data to the visualizer
visualizer.show()           # Finalize and render the figure
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
#OneVsRestClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
clf1 = OneVsRestClassifier(LinearSVC(random_state=0)).fit(X_train, y_train)
#prediction
pred1 = clf1.predict(X_test)
#mean accuracy
score1 = clf1.score(X_test,y_test)


```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, pred1, labels=clf1.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf1.classes_)
disp.plot()
plt.show()
print(classification_report(y_test, pred1))
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
cm = confusion_matrix(y_test, pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()

print(classification_report(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=4, random_state=0)
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
cm = confusion_matrix(y_test, pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()
print(classification_report(y_test, pred))
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
clf = LinearDiscriminantAnalysis()
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
cm = confusion_matrix(y_test, pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()

print(classification_report(y_test, pred))
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(random_state=1,hidden_layer_sizes=2,activation='relu',solver = 'adam', 
                    validation_fraction=0.1, learning_rate_init=0.001, verbose=True,
                    learning_rate='constant', max_iter=300).fit(X_train, y_train)
```

```python
pred = clf.predict(X_test)

cm = confusion_matrix(y_test, pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()
print('test score',clf.score(X_test,y_test))

print(classification_report(y_test, pred))
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
clf = QuadraticDiscriminantAnalysis()
clf.fit(X_train, y_train)

pred = clf.predict(X_test)

cm = confusion_matrix(y_test, pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()
print('test score',clf.score(X_test,y_test))

print(classification_report(y_test, pred))
```
